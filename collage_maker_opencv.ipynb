```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
```

```python
top_right = cv2.imread("aus/top_right.jpg")
top_left = cv2.imread("aus/top_left.jpg")
bottom_right = cv2.imread("aus/bottom_right.jpg")
bottom_left = cv2.imread("aus/bottom_left.jpg")
center = cv2.imread("aus/center.jpg")
```

```python
top_right = cv2.resize(top_right, (200,200))
top_left = cv2.resize(top_left, (200,200))
bottom_right = cv2.resize(bottom_right, (200,200))
bottom_left = cv2.resize(bottom_left, (200,200))
center = cv2.resize(center, (100,100))
```

```python
final = np.zeros((430, 430, 3), np.uint8)
```

```python
final[10:210,10:210] = top_left[:,:]
final[220:420,220:420] = bottom_right[:,:]
final[10:210,220:420] = top_right[:,:]
final[220:420,10:210] = bottom_left[:,:]
final[155:275,155:275] = 0
final[165:265,165:265] = center[:,:]
final = final[:,:,::-1]
cv2.imwrite("Australia.jpg",final[:,:,::-1])
plt.imshow(final)
```

```python
final_csv = np.reshape(final,(430*430,3))
```

```python
df = pd.DataFrame(final_csv)
```

```python
df.to_csv("Australia.csv",index=False,header=("R","G","B"))
```
